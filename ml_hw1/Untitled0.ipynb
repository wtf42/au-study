```python
import cPickle
import numpy as np

with open('mnist.pkl', 'r') as f:
    X_train_raw, Y_train_raw, X_test_raw, Y_test_raw = cPickle.load(f)

X_train = np.array(X_train_raw).astype('float32') / 256
X_test  = np.array(X_test_raw).astype('float32') / 256
Y_train = np.array(Y_train_raw)
Y_test  = np.array(Y_test_raw)

from keras.utils import np_utils
Y_test = np_utils.to_categorical(Y_test)
Y_train = np_utils.to_categorical(Y_train)
```

```python
len(X_train), len(Y_train), len(X_test), len(Y_test), len(X_train[0]), len(Y_train[0])
```

```python
X0 = np.concatenate((X_train, X_test), axis=0)
Y0 = np.concatenate((Y_train, Y_test), axis=0)
from sklearn import cross_validation
X1, X3, Y1, Y3 = cross_validation.train_test_split(X0, Y0, test_size=0.1, random_state=0xA1B2C3D4)
X1, X2, Y1, Y2 = cross_validation.train_test_split(X1, Y1, test_size=0.1/0.9, random_state=0xA1B2C3D4)
print len(X1)
print len(X2)
print len(X3)
X_train = X1; X_test = X2; X_validate = X3;
Y_train = Y1; Y_test = Y2; Y_validate = Y3;
```

```python
from keras.models import Sequential

model = Sequential()
```

```python
from keras.layers import Dense, Activation

model.add(Dense(output_dim=64, input_dim=784))
model.add(Activation("relu"))
model.add(Dense(output_dim=10))
model.add(Activation("softmax"))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
#from keras.optimizers import SGD
#model.compile(loss='categorical_crossentropy', optimizer=SGD(lr=0.01, momentum=0.9, nesterov=True))
```

```python
model.fit(X_train, Y_train, nb_epoch=5, batch_size=32)
```

```python
loss_and_metrics = model.evaluate(X_test, Y_test, batch_size=32)
loss_and_metrics
```
